```python
import pandas as pd

# Load JSON data (Gomla Market - File 1)
df = pd.read_json(r"C:\data analysis\dataset\gomla_facebook_json.json")

# Select required columns only
target_columns = [
    "postId",
    "pageName",
    "text",
    "url",
    "time",
    "timestamp",
    "isVideo",
    "paidPartnership",
    "likes",
    "comments",
    "shares",
    "viewsCount",
    "reactionLikeCount",
    "reactionLoveCount",
    "reactionCareCount",
    "reactionWowCount",
    "reactionHahaCount",
    "reactionSadCount",
    "reactionAngryCount"
]

df_fact = df[target_columns].copy()

# Handle missing values
df_fact["text"] = df_fact["text"].fillna("").astype(str)

# Fill numeric columns with zeros
numeric_cols = [
    "likes", "comments", "shares", "viewsCount",
    "reactionLikeCount", "reactionLoveCount", "reactionCareCount",
    "reactionWowCount", "reactionHahaCount", "reactionSadCount", "reactionAngryCount"
]
df_fact[numeric_cols] = df_fact[numeric_cols].fillna(0)

# Text cleaning
df_fact["cleaned_text"] = df_fact["text"]

# Remove URLs and mentions
df_fact["cleaned_text"] = df_fact["cleaned_text"].str.replace(r"http\S+|www\S+", "", regex=True)
df_fact["cleaned_text"] = df_fact["cleaned_text"].str.replace(r"@\w+", "", regex=True)

# Remove special characters & emojis, preserve Arabic & English
df_fact["cleaned_text"] = df_fact["cleaned_text"].str.replace(r"[^\w\s\u0600-\u06FF]", " ", regex=True)

# Replace underscores
df_fact["cleaned_text"] = df_fact["cleaned_text"].str.replace("_", " ", regex=False)

# Remove extra spaces and convert to lowercase
df_fact["cleaned_text"] = (
    df_fact["cleaned_text"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.lower()
)

# Add project columns
# Key format: FB_GOM_1_ (Facebook_Gomla_File1)
df_fact["PostKey"] = "FB_GOM_1_" + df_fact["postId"].astype(str)

df_fact["Brand"] = "Gomla Market"
df_fact["Platform"] = "Facebook"

# Map video flag to content type
df_fact["ContentType"] = df_fact["isVideo"].map({
    True: "Video",
    False: "Image"
})

# Reorder columns for data warehouse structure
df_fact = df_fact[
    [
        "PostKey",
        "postId",
        "Brand",
        "Platform",
        "pageName",
        "time",
        "timestamp",
        "ContentType",
        "isVideo",
        "paidPartnership",
        "text",
        "cleaned_text",
        "likes",
        "comments",
        "shares",
        "viewsCount",
        "reactionLikeCount",
        "reactionLoveCount",
        "reactionCareCount",
        "reactionWowCount",
        "reactionHahaCount",
        "reactionSadCount",
        "reactionAngryCount",
        "url",
    ]
]

# Export to CSV
output_path = r"C:\data analysis\SocialMedia_Project\Fact_Facebook_Gomla_1.csv"

df_fact.to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print("✅ Fact_Facebook_Gomla_1.csv exported successfully with PostKey.")
```
