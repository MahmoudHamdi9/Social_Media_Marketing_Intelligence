```python
import pandas as pd

# Load TikTok JSON data (Gomla Market)
df_gomla_tt = pd.read_json(r"C:\data analysis\dataset\dataset_tiktok_gomla.json")

# Select required columns
gomla_tt_target_columns = [
    "authorMeta.name",
    "text",
    "diggCount",
    "commentCount",
    "shareCount",
    "playCount",
    "collectCount",
    "videoMeta.duration",
    "musicMeta.musicName",
    "musicMeta.musicAuthor",
    "musicMeta.musicOriginal",
    "createTimeISO",
    "webVideoUrl"
]

df_gomla_tt_fact = df_gomla_tt[gomla_tt_target_columns].copy()

# Handle missing values
df_gomla_tt_fact["text"] = df_gomla_tt_fact["text"].fillna("").astype(str)

# Fill numeric columns with zeros
gomla_tt_numeric_cols = [
    "diggCount",
    "commentCount",
    "shareCount",
    "playCount",
    "collectCount",
    "videoMeta.duration"
]

df_gomla_tt_fact[gomla_tt_numeric_cols] = df_gomla_tt_fact[gomla_tt_numeric_cols].fillna(0)

# Fill music metadata columns
df_gomla_tt_fact["musicMeta.musicName"] = df_gomla_tt_fact["musicMeta.musicName"].fillna("")
df_gomla_tt_fact["musicMeta.musicAuthor"] = df_gomla_tt_fact["musicMeta.musicAuthor"].fillna("")
df_gomla_tt_fact["musicMeta.musicOriginal"] = df_gomla_tt_fact["musicMeta.musicOriginal"].fillna(False)

# Text cleaning
df_gomla_tt_fact["cleaned_text"] = df_gomla_tt_fact["text"]

# Remove URLs
df_gomla_tt_fact["cleaned_text"] = df_gomla_tt_fact["cleaned_text"].str.replace(
    r"http\S+|www\S+",
    "",
    regex=True
)

# Remove mentions
df_gomla_tt_fact["cleaned_text"] = df_gomla_tt_fact["cleaned_text"].str.replace(
    r"@\w+",
    "",
    regex=True
)

# Remove special characters & emojis, preserve Arabic & English
df_gomla_tt_fact["cleaned_text"] = df_gomla_tt_fact["cleaned_text"].str.replace(
    r"[^\w\s\u0600-\u06FF]",
    " ",
    regex=True
)

# Replace underscores
df_gomla_tt_fact["cleaned_text"] = df_gomla_tt_fact["cleaned_text"].str.replace(
    "_",
    " ",
    regex=False
)

# Remove extra spaces
df_gomla_tt_fact["cleaned_text"] = (
    df_gomla_tt_fact["cleaned_text"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

# Convert timestamp to datetime
df_gomla_tt_fact["createTimeISO"] = pd.to_datetime(df_gomla_tt_fact["createTimeISO"])

# Add project columns
df_gomla_tt_fact["PostID"] = "TT_GOM_" + (df_gomla_tt_fact.index + 1).astype(str)

df_gomla_tt_fact["Brand"] = "Gomla Market"

df_gomla_tt_fact["Platform"] = "TikTok"

df_gomla_tt_fact["Source"] = "Apify"

# Reorder columns for data warehouse structure
df_gomla_tt_fact = df_gomla_tt_fact[
    [
        "PostID",
        "Brand",
        "Platform",
        "Source",
        "authorMeta.name",
        "createTimeISO",
        "text",
        "cleaned_text",
        "diggCount",
        "commentCount",
        "shareCount",
        "playCount",
        "collectCount",
        "videoMeta.duration",
        "musicMeta.musicName",
        "musicMeta.musicAuthor",
        "musicMeta.musicOriginal",
        "webVideoUrl"
    ]
]

# Export to CSV
gomla_tt_output_path = r"C:\data analysis\SocialMedia_Project\Fact_Gomla_TikTok.csv"

df_gomla_tt_fact.to_csv(
    gomla_tt_output_path,
    index=False,
    encoding="utf-8-sig"
)

print("✅ Fact_Gomla_TikTok.csv exported successfully with clean gomla_tt variables.")
```
