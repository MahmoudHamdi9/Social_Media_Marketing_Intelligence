```python
import pandas as pd

# Load Instagram JSON data (Gomla Market - File 1)
df_insta = pd.read_json(r"C:\data analysis\dataset\dataset_instagram-gomla.market.json")

# Select required columns for Instagram
insta_target_columns = [
    "id",
    "ownerUsername",
    "caption",
    "productType",
    "url",
    "timestamp",
    "likesCount",
    "commentsCount",
    "videoViewCount",
    "videoPlayCount",
    "videoDuration",
    "hashtags",
    "isPinned"
]

df_insta_fact = df_insta[insta_target_columns].copy()

# Handle missing values
df_insta_fact["caption"] = df_insta_fact["caption"].fillna("").astype(str)

# Fill numeric columns with zeros
insta_numeric_cols = ["likesCount", "commentsCount", "videoViewCount", "videoPlayCount", "videoDuration"]
df_insta_fact[insta_numeric_cols] = df_insta_fact[insta_numeric_cols].fillna(0)

# Fill boolean column
df_insta_fact["isPinned"] = df_insta_fact["isPinned"].fillna(False)

# Text cleaning
df_insta_fact["cleaned_caption"] = df_insta_fact["caption"]

# Remove URLs and mentions
df_insta_fact["cleaned_caption"] = df_insta_fact["cleaned_caption"].str.replace(r"http\S+|www\S+", "", regex=True)
df_insta_fact["cleaned_caption"] = df_insta_fact["cleaned_caption"].str.replace(r"@\w+", "", regex=True)

# Remove special characters & emojis, preserve Arabic & English
df_insta_fact["cleaned_caption"] = df_insta_fact["cleaned_caption"].str.replace(r"[^\w\s\u0600-\u06FF]", " ", regex=True)

# Replace underscores
df_insta_fact["cleaned_caption"] = df_insta_fact["cleaned_caption"].str.replace("_", " ", regex=False)

# Remove extra spaces and convert to lowercase
df_insta_fact["cleaned_caption"] = (
    df_insta_fact["cleaned_caption"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.lower()
)

# Add project columns and generate unique post key
# Key format: IG_GOM_1_ (Instagram_Gomla_File1)
df_insta_fact["PostKey"] = "IG_GOM_1_" + df_insta_fact["id"].astype(str)

df_insta_fact["Brand"] = "Gomla Market"
df_insta_fact["Platform"] = "Instagram"

# Reorder columns for data warehouse structure
df_insta_fact = df_insta_fact[
    [
        "PostKey",
        "id",
        "Brand",
        "Platform",
        "ownerUsername",
        "timestamp",
        "productType",
        "caption",
        "cleaned_caption",
        "likesCount",
        "commentsCount",
        "videoViewCount",
        "videoPlayCount",
        "videoDuration",
        "hashtags",
        "isPinned",
        "url"
    ]
]

# Export to CSV
insta_output_path = r"C:\data analysis\SocialMedia_Project\Fact_Instagram_Gomla_1.csv"

df_insta_fact.to_csv(
    insta_output_path,
    index=False,
    encoding="utf-8-sig"
)

print("✅ Fact_Instagram_Gomla_1.csv exported successfully with unique variables.")
```
